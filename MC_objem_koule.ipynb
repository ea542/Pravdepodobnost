### Výpočet objemu $n$-rozměrné jednotkové koule v metodou Monte Carlo
Označme $I_M$ charakteristickou funkci množiny $M$.

$n$-rozměrná jednotková krychle v $\mathbb R^n$:<br>
$C:=\langle -1, 1\rangle\times\langle -1, 1\rangle\times\cdots\times\langle -1, 1\rangle$

$n$-rozměrná jednotková koule v $\mathbb R^n$:<br>
$K:=\{(x_1, x_2, \ldots, x_n)\in\mathbb R^2:x_1^2+x_2^2+\cdots x_n^2\leq 1\}$

\begin{eqnarray}
S 
&:=& \idotsint_K 1 dx_1 dx_2 \cdots dx_n\\
&=& \idotsint_{\mathbb R^n} I_K(x_1, x_2, \ldots, x_n) dx_1 dx_2 \cdots dx_n\\
&=& \idotsint_{\mathbb R^n} 2^nI_K(x_1, x_2, \ldots, x_n)\left(\frac{1}{2^n}I_C(x_1, x_2, \ldots, x_n))\right)dx_1 dx_2 \cdots dx_n\\
&=& \idotsint_{\mathbb R^n} 2^nI_K(x_1, x_2, \ldots, x_n)f_{X_1, X_2, \ldots, X_n}(x_1, x_2, \ldots, x_n)dx_1 dx_2 \cdots dx_n\\
&=& \mathrm E\left(2^nI_K(X_1, X_2, \ldots, X_n)\right),
\end{eqnarray}

kde $(X_1, X_2, \ldots, X_n)\sim U\left(C\right)$, a tedy $f_{X_1, X_2, \ldots, X_n}(x_1, x_2, \ldots, x_n) = \frac{1}{2^n}I_C(x_1, x_2, \ldots, x_n)$.

Pro nezávislé $(X^{(1)}_1, \ldots, X^{(1)}_n),\ldots,(X^{(N)}_1, \ldots, X^{(N)}_n)\sim U\left(C\right)$ 
označme $$S_N:=\frac{1}{N}\sum_{i=1}^N 2^nI_K(X^{(i)}_1, \ldots, X^{(i)}_n).$$

Podle zákona velkých čísel platí
$$S_N\xrightarrow{sj} E(2^nI_K(X_1, X_2, \ldots, X_n))=S.$$


```python
# načtení modulů
import numpy as np                   # balík pro práci s poli 
from scipy import stats              # modul pro práci s rozděleními náhodných veličin 
from matplotlib import pyplot as plt # modul pro práci s grafy
```

```python
N = 10**7 # počet náhodných vzorků
n = 10 # dimenze
X = np.random.uniform(-1, 1, size=(N, n)) # matice náhodných vzorků (x_i, y_i) ~ U(<0,1> x <0,1>); v řádcích
s = (2**n)*(np.sum(X**2, axis=1)<=1) # pole vzorků
S = np.mean(s) # 4 x výběrový průměr (odhad střední hodnoty)
print("S \u2248 {:.6f}".format(S)) 
```

Podle centrální limitní věty $S_N\dot\sim\mathcal N(E(S_N), D(S_N))$.

```python
DEst = np.var(s) # výběrový rozptyl (odhad rozptylu) veličiny 4I_K
sd_N = (DEst/N)**0.5 # odhad směrodatné odchylky výběrového průměru S_N
```

```python
# interval pro zvolenou spolehlivost 1-alpha
alpha = 0.01
D = stats.norm.ppf(1-alpha/2, loc=0, scale=sd_N) # polovina délky intervalu
print("P(S \u2208 <{:.6f}, {:.6f}>) \u2248 {:.3g}".format(S-D, S+D, 1-alpha))
```

Přesná hodnota:

```python
from scipy.special import gamma
```

```python
# objem n-rozměrné koule
V = np.pi**(n/2)/gamma(n/2+1)

print("Objem {}-rozměrné jednotkové koule: {}".format(n, V))
```
